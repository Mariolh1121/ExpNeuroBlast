Análisis del Proyecto: Cresta Neural y MOXD1

El dataset se centra en el gen MOXD1 y su rol en el desarrollo embrionario en la cresta neural del tronco de Gallus gallus (pollo). Este gen ha sido implicado como un supresor tumoral en neuroblastoma, un tipo de cáncer infantil. En este estudio, utilizaron una técnica de knockout condicional de MOXD1 en células de la cresta neural utilizando CRISPR/Cas9. Los datos de expresión génica obtenidos mediante secuenciación de alto rendimiento permiten evaluar cómo afecta la eliminación de este gen al desarrollo y a la expresión de otros genes importantes para la diferenciación celular.
